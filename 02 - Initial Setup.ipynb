Rank the arch mirror servers by speed

```python
def rankmirrors():
    """ Take the original mirror list file and generate the fastest mirror list """
    with tempfile.TemporaryDirectory() as tempdir:
        temp_mirrorlist = os.path.join(tempdir, 'mirrorlist')

        in_block = False
        with open(PACMAN_MIRRORLIST, 'rt') as fp:
            with open(temp_mirrorlist, 'wt') as gp:
                
                for line in fp:
                    line = line.strip()
                    
                    if not line.startswith('#'):
                        gp.write(line + os.linesep)
                        continue

                    if line.startswith('##'):
                        if 'United States' in line:
                            in_block = True
                        else:
                            in_block = False

                    if line.startswith('#Server') and in_block:
                        gp.write(line[1:] + os.linesep)
                    else:
                        gp.write(line + os.linesep)
        out = subprocess.check_output(['rankmirrors', '-n', '6', temp_mirrorlist]).decode('utf-8')
        with open(temp_mirrorlist, 'wt') as fp:
            fp.write(out)
        print(open(temp_mirrorlist, 'rt').read())
rankmirrors()
```

Bash Script to Drive the Initial Config

```python
%%bash

# Toast the old partition table and write a single new one
sgdisk --clear /dev/sda

sgdisk -n 1:0:2097151 -c 1:"Linux Boot" -t 1:8300 /dev/sda
sgdisk -n 2:2097152:0 -c 2:"Linux" -t 2:8300 /dev/sda

# Partition the Disks
mkfs -t ext4 /dev/sda1

cryptsetup luksFormat --hash sha256 --cipher aes-xts-essiv:sha256 /dev/sda2
cryptsetup luksOpen /dev/sda2 root

mkfs -t btrfs /dev/mapper/root

# Mount all the disks
mkdir -p /mnt
mount -t btrfs /dev/mapper/root /mnt

mkdir -p /mnt/boot
mount -t ext4 /dev/sda1 /mnt/boot
```

```python
%%bash

# Bootstrap pacman
pacstrap -i /mnt base base-devel btrfs-progs zsh

genfstab -U -p /mnt >> /mnt/etc/fstab

arch-chroot /mnt /bin/zsh

pacman -S syslinux gptfdisk mtools
pacman -S dialog iw wpa_supplicant wpa_actiond dhcpd
pacman -S sudo vim git python ipython 
```
